```python
#Separations des variables quantitatives et qualitatives
def split_data(df):
    cat_data = []
    num_data = []
    for i,c in enumerate(df.dtypes):
        #df.dtypes
        if c==object:
            cat_data.append(df.iloc[:,i])
        else:
            num_data.append(df.iloc[:,i])

    return pd.DataFrame(cat_data).transpose(), pd.DataFrame(num_data).T
```

```python
def encoding_cat(df):
    # ordinal variable
    code = {"Basic":0,
            "2n Cycle":1,
            "Master": 2,
            "PhD":3,
            "Graduation":4
            }
    var = ['Education']
    for col in var:
        df.loc[:,col] = df[col].map(code)
        
    # nominal variable
    var_without_grad = ['Marital_Status']
    df_1 = df[var_without_grad]
    df_1 = pd.get_dummies(df_1, drop_first=True)
    df = df.drop(var_without_grad, axis='columns')

    df = pd.concat([df, df_1], axis='columns')
 
    return df
```

```python
# cat_data.head(6)
cat_data = encoding_cat(cat_data)
```

* On codifie les variables qualitatives
* Certaines variables sont qualitatives ordinale dont nous concervons ce ordre dans la normalisation
* D'autres sont qualitatives norminales

```python
cat_data.head()
```

```python
# Feature Scaling
rs = RobustScaler()
num_data = pd.DataFrame(rs.fit_transform(num_data), columns=num_data.columns)
num_data.head()
```

* On normalise les variables quantitatives avec du Robust Scaler parce que c'est moins sensible aux valeurs extrèmes

```python
df_mod = pd.concat([cat_data, num_data], axis='columns')
df_mod.head(5)
```

* Enfin nous mettons tous le dataset en forme dans df_mod pour faire la modélisation après 

```python
df_mod = df_mod.drop('Tranche d\'ages',axis=1)
df_mod.head()
```
